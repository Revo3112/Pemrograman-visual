```python
# Melakukan import terhadap library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt

# Melakukan gambar sebuah bendera dengan 
# 1. Warna atas biru gelap dengan RGB rgba(50,50,100,255), col 300 - 1100, row 200 - 300
# 2. Warna bawah biru muda dengan RGB rgba(80,80,200,255), col 300 - 1100, row 300 - 600

# Memmbuat ukuran canvas dengan bacground hitam 
col = 1400
row = 800

# Menyiapkan background canvas dengan ukuran yang sudah di tetapkan
bcg_canvas = np.zeros(shape=(row, col, 3), dtype=np.uint8)

# Menyiapkan posisi titik yang akan di iterasi
col_atas_dan_bawah = [300, 1100]
row_atas = [200, 300]
row_bawah = [300, 600]

# Melakukan iterasi untuk kolom, row dan mengganti warna piksel bagian atas sesuai dengan kriteria
for i in range(col_atas_dan_bawah[0], col_atas_dan_bawah[1]):
    for j in range(row_atas[0], row_atas[1]):
        # Melakukan setting terhadap saluran 1 yaitu red
        bcg_canvas[j, i, 0] = 50
        # Melakukan setting terhadap saluran 2 yaitu green
        bcg_canvas[j, i ,1] = 50
        # Melakukan setting terhadap saluran 3 yaitu blue
        bcg_canvas[j, i, 2] = 100

# Melakukan iterasi untuk kolom, row dan mengganti warna piksel bagian bawah sesuai dengan kriteria
for i in range(col_atas_dan_bawah[0], col_atas_dan_bawah[1]):
    for j in range(row_bawah[0], row_bawah[1]):
        # MElakukan setting terhadap saluran 1 yaitu red
        bcg_canvas[j, i, 0] = 80
        bcg_canvas[j, i, 1] = 80
        bcg_canvas[j, i, 2] = 200

# Melakukan pengaturan ukran dari plot yang akan dbuat
plt.figure(figsize=(5, 5))        
# Menampilkan gambar yang telah di buat
plt.imshow(bcg_canvas)
```

```python
# Melakukan Import Library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt

# Menentukan jarak linespace yang akan digunakan sebagai nilai x
x = np.linspace(-9, 9, 100000)

# Melakukan pengaturan ukuran dari plot yang akan dibuat 
plt.figure(figsize=(5, 5))

# Menentukan persamaan matematika yang diinginkan 
# persamaan lingkaran x^2 + y^2 = r^2
# persamaan lingkaran (x-a)^2 + (y-b)^2 = r^2
# persamaan lingkaran y = (r^2 - x^2)^0.5

# Membuat lingkaran ditengah dengan radius 5 dengan titik pusat (0, 0)
y1 = (5**2 - x**2)**0.5
y2 =  -(5**2 - x**2)**0.5

# Membuat lingkaran yang mengelilinginya 1 dengan radius 2 dengan titik pusat (-6, 0)
y3 = (2**2 - (x - (-6))**2)**0.5
y4 = -(2**2 - (x - (- 6))**2)**0.5

# Membuat lingkaran yang mengelinginya 2 dengan radius 2 dengan titik pusat (-4, 4)
y5 = 4 + (2**2 - (x - (-4))**2)**0.5
y6 = 4 -(2**2 - (x - (-4))**2)**0.5

# Membuat lingkaran yang mengelinginya 3 dengan radius 2 dengan titik pusat (0, 6)
y7 = 6 + (2**2 - (x - 0)**2)**0.5
y8 = 6 -(2**2 - (x - 0)**2)**0.5

# Membuat lingkaran yang mengelinginya 4 dengan radius 2 dengan titik pusat (4, 4)
y9 = 4 + (2**2 - (x - 4)**2)**0.5
y10 = 4 -(2**2 - (x - 4)**2)**0.5

# Membuat lingkaran yang mengelinginya 5 dengan radius 2 dengan titik pusat (6, 0)
y11 =  (2**2 - (x - 6)**2)**0.5
y12 = -(2**2 - (x - 6)**2)**0.5

# Membuat lingkaran yang mengelinginya 6 dengan radius 2 dengan titik pusat (4, -4)
y13 = -4 +  (2**2 - (x - 4)**2)**0.5
y14 = -4 - (2**2 - (x - 4)**2)**0.5

# Membuat lingkaran yang mengelinginya 7 dengan radius 2 dengan titik pusat (0, -6)
y15 = -6 + (2**2 - (x)**2)**0.5
y16 = -6 - (2**2 - (x)**2)**0.5

# Membuat lingkaran yang mengelinginya 8 dengan radius 2 dengan titik pusat (-4, -4)
y17 = -4 + (2**2 - (x - (-4))**2)**0.5
y18 = -4 - (2**2 - (x - (-4))**2)**0.5

# Melakukan plotting dari persamaan yang telah dibuat dengan warna yang sudah ditentukan 
# warna lingkaran pusat di isi dengan rgba(255,191,205,255)
# warna lingkaran pinggir di isi dengan rgba(253,2,251,255)
# warna tepi lingkaran tengah di isi dengan rgba(201,51,62,255)
# warnaa tepi lingkaran pinggir di isi dengan rgba(125,121,199,255)
plt.plot(x, y1, color=(201/255, 51/255, 62/255), linewidth=2.5) 
plt.plot(x, y2, color=(201/255, 51/255, 62/255), linewidth=2.5)
plt.plot(x, y3, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y4, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y5, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y6, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y7, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y8, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y9, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y10, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y11, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y12, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y13, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y14, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y15, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y16, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y17, color=(125/255, 121/255, 199/255), linewidth=2.5)
plt.plot(x, y18, color=(125/255, 121/255, 199/255), linewidth=2.5)

# Melakukan pengisiian warna tengah dari lingkaran yang telah dibuat
plt.fill_between(x, y1, y2, color=(255/255, 191/255, 205/255))
plt.fill_between(x, y3, y4, color=(253/255, 2/255, 251/255))
plt.fill_between(x, y5, y6, color=(253/255, 2/255, 251/255))
plt.fill_between(x, y7, y8, color=(253/255, 2/255, 251/255))
plt.fill_between(x, y9, y10, color=(253/255, 2/255, 251/255))
plt.fill_between(x, y11, y12, color=(253/255, 2/255, 251/255))
plt.fill_between(x, y13, y14, color=(253/255, 2/255, 251/255))
plt.fill_between(x, y15, y16, color=(253/255, 2/255, 251/255))
plt.fill_between(x, y17, y18, color=(253/255, 2/255, 251/255))

# Melakukan setting terhadap axis agar terlihat simetris
plt.axis('equal')
# Menampilkan legend terhadap plot yang telah dibuat
plt.legend(['bunga'], loc='upper left')
# Menampilkan grid agar terlihat jelas
plt.grid()
# Menampilkan plot yang telah dibuat
plt.show()
```
