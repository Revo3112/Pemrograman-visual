```python
# Membuat garis lurus diketahui titik a (100, 200) dan b(800, 600) dilewati oleh garis g
# Rumus garis lurus bila diketahui dua titik 
# y = m(x-x1) + y1
# my = (y2 - y1) / (x2 - x1)
# y = (y2 - y1) / (x2 - x1) * (x-x1) + y1

import matplotlib.pyplot as plt
import numpy as np

radius = 10
col = 1000
row = 1000
bcg_canvas = np.zeros((row, col, 3), dtype=np.uint8)
bcg_canvas[:,:,:] = 0

def rumus_persamaan_garis(x1, y1, x2, y2):
    def persamaan(x):
        m = (y2 - y1) / (x2 - x1)
        y = m * (x - x1) + y1
        return y
    return persamaan

def membuat_lingkaran(x, y, radius, warna, gambar, row, col):
    for j in range(max(int(y - radius), 0), min(int(y + radius), row)):
        for i in range(max(int(x-radius), 0), min(int(x+radius), col)):
            if ((i-x)**2 + (j-y)**2 < radius**2):
                gambar[j,i] = warna
    return gambar

titik_a = (100, 200)
titik_b = (800, 600)

# Gambar dua titik tujuan 
membuat_lingkaran(titik_a[0], titik_a[1], 20, [213,85,239], bcg_canvas, row, col)
membuat_lingkaran(titik_b[0], titik_b[1], 20, [155,0,186], bcg_canvas, row, col)

for x in range(max(titik_a[0] - 50, 0), min(titik_b[0] + 50, col)):
    y = rumus_persamaan_garis(titik_a[0], titik_a[1], titik_b[0], titik_b[1])
    warna = [int(255 * x / col), 255, 255]  # RGB color
    membuat_lingkaran(x, int(y(x)), radius, warna, bcg_canvas, row, col)

#Set lebar canvas sesuai dengan resolusi
plt.figure(figsize=(20,20))
plt.imshow(bcg_canvas)
plt.show()
```

```python
# Membuat garis lurus diketahui titik a (100, 200) dan b(1800, 200) dilewati oleh garis g
# set  col row baru 
col = 1920
row = 1200
# reset canvas 
bcg_canvas = np.zeros((row, col, 3), dtype=np.uint8)
bcg_canvas[:,:,:] = 0

titik_a = (100, 200)
titik_b = (1800, 200)

# Gambar dua titik tujuan
membuat_lingkaran(titik_a[0], titik_a[1], 20, [213,85,239], bcg_canvas, row, col)
membuat_lingkaran(titik_b[0], titik_b[1], 20, [155,0,186], bcg_canvas, row, col)

for x in range(max(titik_a[0] - 50, 0), min(titik_b[0] + 50, col)):
    y = rumus_persamaan_garis(titik_a[0], titik_a[1], titik_b[0], titik_b[1])
    warna = [255, int(255 * (1 - x / col)), 255]  # RGB color
    membuat_lingkaran(x, int(y(x)), radius, warna, bcg_canvas, row, col)

#Set lebar canvas sesuai dengan resolusi
plt.figure(figsize=(20,20))
plt.imshow(bcg_canvas)
plt.show()
```

```python
# Membuat garis vertikal dari titik a (100, 200) ke titik b(100, 1000)
# Gradien dari y 
# my = (x2 - x1) / (y2 - y1)
# y = (x2 - x1) / (y2 - y1) * (y-y1) + x1

def garis_vertikal(x1, y1, x2, y2):
    def persamaan(y):
        m = (x2 - x1) / (y2 - y1)
        x = m * (y - y1) + x1
        return x
    return persamaan

# set  col row baru
col = 1000
row = 1200
# reset canvas
bcg_canvas = np.zeros((row, col, 3), dtype=np.uint8)
bcg_canvas[:,:,:] = 0

titik_a = (100, 200)
titik_b = (100, 1000)

# Gambar dua titik tujuan
membuat_lingkaran(titik_a[0], titik_a[1], 20, [213,85,239], bcg_canvas, row, col)
membuat_lingkaran(titik_b[0], titik_b[1], 20, [155,0,186], bcg_canvas, row, col)

for y in range(max(titik_a[1] - 50, 0), min(titik_b[1] + 50, row)):
    x = garis_vertikal(titik_a[0], titik_a[1], titik_b[0], titik_b[1])
    warna = [255, 255, int(255 * (1 - y / row))]  # RGB color
    membuat_lingkaran(int(x(y)), y, radius, warna, bcg_canvas, row, col)

#Set lebar canvas sesuai dengan resolusi
plt.figure(figsize=(20,20))
plt.imshow(bcg_canvas)
plt.show()
```
