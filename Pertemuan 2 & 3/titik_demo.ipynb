```python
print("\033c")

import numpy as np
import matplotlib.pyplot as plt

# Tentukan koordinat 2 buah titik 
koordinat_x = [200, 800]
koordinat_y = [200, 800]

# The user decide line width 
lw = int (10)

# calculate the half of the line width
half_lw = int(lw / 2)

# Setting the size of the canvas 
col = 900
row = 900
print("col = ",col,"row = ", row)

# Preparing the black canvas dengan menggunakan metod  manipulasi array pixel gambar pada pengolahan citra (computer vision)
bcg_canva = np.zeros((row, col, 3), dtype=np.uint8)
bcg_canva[:,:,:] = 0

# Menggambarkan titik dalam satu piksel 
bcg_canva[koordinat_y[0], koordinat_x[0], 0] = 123
bcg_canva[koordinat_y[1], koordinat_x[1], 0] = 123

# Membuat lingkaran agar jadi lebih besar 
for i in range(koordinat_x[0] - lw, koordinat_x[0] + lw):
    for j in range(koordinat_y[0] - lw, koordinat_y[0] + lw):
        if (i - koordinat_x[0])**2 + (j - koordinat_y[0])**2 <= half_lw**2:
            bcg_canva[j, i, 0] = 123

for i in range(koordinat_x[1] - lw, koordinat_x[1] + lw):
    for j in range(koordinat_y[1] - lw, koordinat_y[1] + lw):
        if (i - koordinat_x[1])**2 + (j - koordinat_y[1])**2 <= half_lw**2:
            bcg_canva[j, i, 0] = 123

plt.imshow(bcg_canva)
plt.show()
```

```python

# Mencari persamaan garis
def persamaan_garis(x1, y1, x2, y2):
    def garis(x):
        return ((y2 - y1) / (x2 - x1)) * (x - x1) + y1

    return garis

y = persamaan_garis(koordinat_x[0], koordinat_y[0], koordinat_x[1], koordinat_y[1])

# Membuat garis
x_values  = np.linspace(koordinat_x[0], koordinat_x[1], 1000)  # 100 titik antara titik awal dan akhir
y_values = [int(y(x)) for x in x_values]  # Mengonversi nilai y menjadi integer

# Menggambarkan garis dengan menggunakan metode manipulasi array pixel gambar pada pengolahan citra (computer vision)
for i in range(len(x_values)):
    x = int(x_values[i])
    y = y_values[i]
    for j in range(y - lw, y + lw):
        for k in range(x - lw, x + lw):
            if (k - x)**2 + (j - y)**2 <= half_lw**2:  # Menggunakan lingkaran untuk menggambar piksel
                bcg_canva[j, k, 0] = 123

plt.imshow(bcg_canva)
plt.show()
```

```python
# Membuat dua titik vertikal 
cor_x = [500, 500]
cor_y = [100, 800]

# Menggambarkan titik dalam satu piksel
bcg_canva[cor_y[0], cor_x[0], 2] = 123
bcg_canva[cor_y[1], cor_x[1], 2] = 123

# Membuat lingkaran agar jadi lebih besar
for i in range(cor_x[0] - lw, cor_x[0] + lw):
    for j in range(cor_y[0] - lw, cor_y[0] + lw):
        if (i - cor_x[0])**2 + (j - cor_y[0])**2 <= half_lw**2:
            bcg_canva[j, i, 2] = 123

for i in range(cor_x[1] - lw, cor_x[1] + lw):
    for j in range(cor_y[1] - lw, cor_y[1] + lw):
        if (i - cor_x[1])**2 + (j - cor_y[1])**2 <= half_lw**2:
            bcg_canva[j, i, 2] = 123

plt.imshow(bcg_canva)
plt.show()
```

```python
# rumus vertikal
def garis_vertikal(range_awal, range_akhir, x, jumlah_lingkaran):
    for i in range(range_awal, range_akhir):
        for n in range(jumlah_lingkaran):
            y = range_awal + (i - range_awal) / jumlah_lingkaran * n  # Hitung posisi y dari setiap lingkaran
            for j in range(x - lw, x + lw):
                if (j - x)**2 + (i - y)**2 <= half_lw**2:
                    bcg_canva[int(i), int(j), 2] = 123

range_awal = cor_y[0]
range_akhir = cor_y[1]
jumlah_lingkaran = 1000  # Jumlah lingkaran di antara dua titik

garis_vertikal(range_awal, range_akhir, cor_x[0], jumlah_lingkaran)

plt.imshow(bcg_canva)
plt.show()
```

```python
# garis horizontal
cor_x = [100, 800]
cor_y = [500, 500]

# Menggambarkan titik dalam satu piksel
bcg_canva[cor_y[0], cor_x[0], 1] = 123
bcg_canva[cor_y[1], cor_x[1], 1] = 123

# Membuat lingkaran agar jadi lebih besar
for i in range(cor_x[0] - lw, cor_x[0] + lw):
    for j in range(cor_y[0] - lw, cor_y[0] + lw):
        if (i - cor_x[0])**2 + (j - cor_y[0])**2 <= half_lw**2:
            bcg_canva[j, i, 1] = 123

for i in range(cor_x[1] - lw, cor_x[1] + lw):
    for j in range(cor_y[1] - lw, cor_y[1] + lw):
        if (i - cor_x[1])**2 + (j - cor_y[1])**2 <= half_lw**2:
            bcg_canva[j, i, 1] = 123

plt.imshow(bcg_canva)
plt.show()
```

```python
# membuat garis Horizontal

def garis_horizontal(range_awal, range_akhir, y, jumlah_lingkaran):
    for i in range(range_awal, range_akhir):
        for n in range(jumlah_lingkaran):
            x = range_awal + (i - range_awal) / jumlah_lingkaran * n  # Hitung posisi x dari setiap lingkaran
            for j in range(y - lw, y + lw):
                if (j - y)**2 + (i - x)**2 <= half_lw**2:
                    bcg_canva[int(j), int(i), 1] = 123

range_awal = cor_x[0]
range_akhir = cor_x[1]
jumlah_lingkaran = 1000  # Jumlah ling

garis_horizontal(range_awal, range_akhir, cor_y[0], jumlah_lingkaran)

plt.imshow(bcg_canva)
plt.show()
```
